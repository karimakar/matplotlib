```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# inputs to plotting functions
np.random.seed(12536584)

data = {

"a" : np.arange(50),
"c":np.random.randint(0,50,50),
"d":np.random.randn(50)}
data['b'] = data["a"] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

fig,ax = plt.subplots(figsize=(10,5.2),layout='constrained')
ax.scatter("a","b",c="c",s="d",data=data)
ax.set_xlabel(' entry a')
ax.set_ylabel(' entry b')


```

```python
# coding styles
x = np.linspace(0,2,100)
fig,ax = plt.subplots(figsize=(10,5.3),layout='constrained')
ax.plot(x,x,label='linear')
ax.plot(x,x**2,label='quadratic')
ax.plot(x,x**3,label='cubic')
ax.plot(x,np.log(x),label="log")
ax.plot(x,np.exp(x),label='exp')
ax.plot(x,np.sqrt(x),label='sqrt')
ax.set_xlabel('x input')
ax.set_ylabel("x out")
ax.legend()
```

```python
# using helper function
def helper(ax,data1,data2,param_dic):
    '''
    a helper fubction to make graph
    '''
    out = ax.plot(data1,data2,**param_dic)
    return out
```

```python
data1,data2,data3,data4=np.random.randn(4,100)
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(10,5.2))
helper(ax1,data1,data2,{'marker':'o'})
helper(ax2,data3,data4,{'marker':'x'})
```

```python
# color
fig,ax=plt.subplots(figsize=(12,5.2))
ax.scatter(data1,data2,s=90,facecolor='red',edgecolor="k")
```

```python
#linewidth,linestye,and markersize
fig,ax = plt.subplots(figsize=(12,6.3))
ax.plot(data1,'o',label='data1')
ax.plot(data2,'h',label='data2')
ax.plot(data3,'v',label='data3')
ax.plot(data4,'d',label='data4')
ax.legend()

```
